# 2-1 加载图像

```python
import cv2 as cv
import numpy as np
print(cv.__version__)
```

**imread(file, mode)**
+ cv.IMREAD_COLOR： 加载彩色图像。任何图像的透明度都会被忽视。它是默认标志 / 1
+ cv.IMREAD_GRAYSCALE：以灰度模式加载图像 / 0
+ cv.IMREAD_UNCHANGED：加载图像，包括alpha通道 / -1

```python
img = cv.imread("images\image1.png")
```

**imshow(window_name, object)**

**waitKey(duration_miliseconds)**
1. duration == 0:无限期等待用户按键，按下后继续执行后续语句
2. duration != 0:在duration内等待用户按键，若在时间段内按下则返回该键的ascii值；若未按下则返回-1

有窗口存在时waitKey才会有效

如果使用的是64位计算机，则必须k = cv.waitKey(0)按如下所示修改行：k = cv.waitKey(0) & 0xFF

**destroyAllWindows()**

```python
cv.imshow("window", img)
k = cv.waitKey(0)
if k == 27:         # 等待ESC退出
    cv.destroyAllWindows()
elif k == ord('s'): # 等待关键字，保存和退出
    cv.imwrite('PickeyBlinders.png',img)
    print("图像已保存")
    cv.destroyAllWindows()
```

# 3-1 图像的基本操作

1. Numpy访问图像像素

可以通过二维数组（横纵坐标）来表示图片的某一个像素，对于RGB图像，返回值是一个红绿蓝的三维数组

也可以通过三维数组来访问，第三个参数为要筛选的颜色（蓝0/绿1/红2）

```python
px = img[100, 100]
print(px)
```

```python
blue = img[100, 100, 0]
print(blue)
```

```python
img[:, :, 2] = 0 # 将所有红色像素设为0
```

2. cv访问图像像素 & 图像属性



```python
img.item(100, 100, 2) #访问红色像素
```

```python
img.itemset((100,100,2), 40) #修改像素
img.item(100, 100, 2)
```

```python
print(img.shape) # 行，列，通道数
print(img.size) # 像素总数
print(img.dtype) # 数据类型
```

```python
# 编辑图片某一块区域内的所有像素
block = img[280:340, 330:390]
img[273:333, 100:160] = block
```
